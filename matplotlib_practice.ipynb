```python
%matplotlib inline
```

```python
## Import Dependencies 
import matplotlib.pyplot as plt
```

```python
### Set the x axis to a list of strings for each month 
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

## Set the y axis to a list of floats as the total fare in US dollars for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot 
plt.plot(x_axis, y_axis)
```

```python
# Create a plot with ax.plt
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create a Plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color= "Red", linewidth=2, label="Boston")
# Create lavels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set limit between 0 and 45
plt.ylim(0, 45)
plt.title("PyBer Fare by Month")
plt.legend()
plt.grid()
```

```python
## Skill Drill
ax.plot(x_axis, y_axis, marker="", color="Green", linewidth=2, label="Boston")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_ylim(0,45)
ax.grid()
ax.set_title("PyBer Fare by Month")
ax.legend()

```

```python
## Bar Chart
plt.bar(x_axis, y_axis)
```

```python
## Annotate Bar Chart
plt.bar(x_axis, y_axis, color="Green", label="Boston")
# Create labels for the x and y axes 
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a Title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
## Horizontal Bar Chart
plt.barh(x_axis, y_axis)
```

```python
### Data on opposite axes? 
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
plt.barh(x_axis, y_axis, color="Magenta", label="Boston")
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
## Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
ax.barh(x_axis, y_axis, color="Cyan", label="Chicago")
ax.legend()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

```

```python
# Scatter plot with MATLAB
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(y_axis, x_axis, marker="d", color="Red", label="Chicago")
plt.legend()
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.xlim(0,50)
plt.gca().invert_yaxis()

```

```python
## Create a Bubble Chart
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
## Pie Charts
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct="%.1f%%")
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
## Pie Chart using Object Oriented interface
fig, ax= plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```
